```python
import pandas as pd
```

```python
with open('yelp_academic_dataset_review.json', 'rb') as f:
    data= f.readlines()
```

```python
data1= list()
```

```python
import ast
```

```python
for i in range(5000):
    data_temp= data[i].decode('utf-8')
    data_temp= ast.literal_eval(data_temp)
    data1.append(data_temp['text'])
```

```python
good_reviews= list()
```

```python
okay_reviews= list()
```

```python
for i in range(10000):
    data_temp = data[i].decode('utf-8')
    data_temp= ast.literal_eval(data_temp)
    stars= data_temp['stars']
    if stars>3:
        good_reviews.append(data_temp['text'])
    else:
        okay_reviews.append(data_temp['text'])
```

```python
from nltk.stem.wordnet import WordNetLemmatizer
```

```python
nltk.download()
```

```python
import string
```

```python
from nltk.corpus import stopwords
```

```python
stop= set(stopwords.words('english'))
```

```python
exclude= set(string.punctuation)
```

```python
lemma= WordNetLemmatizer()
```

```python
def clean(doc):
    stop_free= " ".join([i for i in doc.lower().split() if i not in stop])
    punc_free= "".join(ch for ch in stop_free if ch not in exclude)
    normalized = " ".join(lemma.lemmatize(word) for word in punc_free.split())
    return normalized
```

```python
doc_clean= [clean(doc).split() for doc in data1] 
```

```python
doc_clean_good= [clean(doc).split() for doc in good_reviews]
```

```python
doc_clean_okay = [clean(doc).split() for doc in okay_reviews]
```

```python
import gensim
```

```python
from gensim import corpora
```

```python
dictionary= corpora.Dictionary(doc_clean)
```

```python
doc_term_matrix = [dictionary.doc2bow(doc) for doc in doc_clean]
```

```python
Lda= gensim.models.ldamodel.LdaModel
```

```python
dictionary_good= corpora.Dictionary(doc_clean_good)
```

```python
dictionary_okay = corpora.Dictionary(doc_clean_okay)
```

```python
doc_term_matrix_good= [dictionary_good.doc2bow(doc) for doc in doc_clean_good]
```

```python
doc_term_matrix_okay= [dictionary_okay.doc2bow(doc) for doc in doc_clean_okay]
```

```python
ldamodel= Lda(doc_term_matrix, num_topics=50, id2word= dictionary, passes=20)
```

```python
print(ldamodel.print_topics(num_topics=50, num_words=10))
```

```python
ldamodel_good= Lda(doc_term_matrix_good, num_topics=50, id2word= dictionary_good, passes=20)
```

```python
ldamodel_okay= Lda(doc_term_matrix_okay, num_topics=50, id2word= dictionary_okay, passes=20)
```
